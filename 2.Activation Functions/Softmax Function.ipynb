```python
import numpy as np
import matplotlib.pyplot as plt

# Define the softmax function
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Numerical stability
    return exp_x / np.sum(exp_x)

# Generate x values
x = np.linspace(-5, 5, 100)  # Input range from -5 to 5

# Create multiple class scores
scores = np.vstack([x, np.ones_like(x), 0.5 * np.ones_like(x)])

# Apply softmax to each set of scores
softmax_values = np.array([softmax(score) for score in scores.T]).T

# Plot the softmax probabilities
for i, probs in enumerate(softmax_values):
    plt.plot(x, probs, label=f'Class {i+1}')

plt.title('Softmax Function')
plt.xlabel('Input Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
