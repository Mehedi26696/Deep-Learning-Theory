```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python

# Define the softplus function
def softplus(x):
    return np.log(1 + np.exp(x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Apply softplus function
y = softplus(x)

# Plot the softplus function
plt.plot(x, y, label='Softplus', color='b')
plt.title('Softplus Activation Function')
plt.xlabel('x')
plt.ylabel('softplus(x)')
plt.legend()
plt.grid(True)
plt.show()
```
