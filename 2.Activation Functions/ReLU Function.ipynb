```python
def relu(x):
    return max(0,x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function using NumPy
def relu(x):
    return np.maximum(0, x)

# Generate x values
x = np.linspace(-100, 10, 100)

# Plot the ReLU function
plt.plot(x, relu(x))
plt.title('Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.show()
```
